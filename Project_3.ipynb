```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing the dataset
df = pd.read_csv('Bestsellers with categories.csv')
```

```python
df.head(25)
```

```python
df.tail(25)
```

```python
# Checking for null values
df.isnull().sum()
```

```python
# Finding the value that divides 10% of data of Price
q10 = np.quantile(df['Price'], 0.10)
print("The value that divides 10% of column Price:", q10)
```

```python
# Quantile of column Price
q1_price = np.quantile(df['Price'], 0.25)
q2_price = np.quantile(df['Price'], 0.50)
q3_price = np.quantile(df['Price'], 0.75)

print("Q1 of column Price:", q1_price)
print("Q2 of column Price:", q2_price)
print("Q3 of column Price:", q3_price)

# IQR of column Price
iqr_price = q3_price - q1_price
print("\nIQR of column Price:", iqr_price)
```

```python
# Quantile of column User Rating
q1_user_rating = np.quantile(df['User Rating'], 0.25)
q2_user_rating = np.quantile(df['User Rating'], 0.50)
q3_user_rating = np.quantile(df['User Rating'], 0.75)

print("Q1 of column User Rating", q1_user_rating)
print("Q2 of column User Rating", q2_user_rating)
print("Q3 of column User Rating", q3_user_rating)

# IQR of column User Rating
iqr_user_rating = q3_user_rating - q1_user_rating
print("\nIQR of column User Rating:", iqr_user_rating)
```

```python
# Boxplot for Price
plt.boxplot(df['Price'], vert=False)
plt.title('Boxplot for Price')
plt.xlabel('Price')
plt.show()
```

```python
# Boxplot for Price
plt.boxplot(df['User Rating'], vert=False)
plt.title('Boxplot for User Rating')
plt.xlabel('User Rating')
plt.show()
```
