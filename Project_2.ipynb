```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing the dataset
df = pd.read_csv('Bestsellers with categories.csv')
df.head(20)
```

```python
df.tail(20)
```

```python
# Checking for null values
df.isnull().sum()
```

```python
# Finding the variance and standard deviation of User Rating
var_User_Rating = np.var(df['User Rating'])
print("Variance of User Rating:",var_User_Rating)
std_User_Rating = np.std(df['User Rating'])
print("Standard Deviation of User Rating:",std_User_Rating)
```

```python
# Finding the variance and standard deviation of Price
var_price = np.var(df['Price'])
print("Variance of Price:",var_price)
std_price = np.std(df['Price'])
print("Standard of Price:",std_price)
```

```python
# Checking the distribution of User Rating by creating a histogram
plt.hist(df['User Rating'], bins=14, edgecolor='magenta', color='pink')
plt.title('Distribution of User Rating')
plt.xlabel('User Rating')
plt.ylabel('Frequency')
```

```python
# Checking the distribution of User Rating by creating a histogram
plt.hist(df['Price'], bins=14, edgecolor='indigo', color='purple')
plt.title('Distribution of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
```
