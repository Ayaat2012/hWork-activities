```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing the dataset
df = pd.read_csv("Bestsellers with categories.csv")
```

```python
# First 30 rows
df.head(30)
```

```python
# Last 30 rows
df.tail(30)
```

```python
# Checking for null values
df.isnull().sum()
```

```python
# Finding the frquency of categories for Genre
df['Genre'].value_counts()
```

```python
# Finding the median value of Genre
genre_cat = ["Non Fiction", "Fiction"]
df["Genre"] = pd.Categorical(df["Genre"], categories=genre_cat, ordered=True)
median_index = np.median(df["Genre"].cat.codes)
median_value = df["Genre"].cat.categories[int(median_index)]
print("Median of column Genre:", median_value)
```

```python
# Creating a bar chart for Genre
plt.figure(figsize=(9, 5))
sns.countplot(data=df, x='Genre', palette='summer')
plt.title('Distribution of Book Genres')
plt.xlabel('Genre', fontsize = 17)
plt.ylabel('Count', fontsize = 17)
plt.xticks(fontsize = 17,rotation=45)
plt.show()
```
