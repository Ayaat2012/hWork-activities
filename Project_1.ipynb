```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import statistics as stats
```

```python
# Reading the dataset
df = pd.read_csv('Bestsellers with categories.csv')
df.head(20)
```

```python
# Checking for nul values
df.isnull().sum()
```

```python
# Finding the mean, mode and median of User Rating
print(f'Mean: {stats.mean(df["User Rating"])}')
print(f'Mode: {stats.mode(df["User Rating"])}')
print(f'Median: {stats.median(df["User Rating"])}')
```

```python
# Finding the mean, mode and median of Price
print(f'Mean: {stats.mean(df["Price"])}')
print(f'Mode: {stats.mode(df["Price"])}')
print(f'Median: {stats.median(df["Price"])}')
```

```python
# Finding the mean, mode and median of Reviews
print(f'Mean: {stats.mean(df["Reviews"])}')
print(f'Mode: {stats.mode(df["Reviews"])}')
print(f'Median: {stats.median(df["Reviews"])}')
```
