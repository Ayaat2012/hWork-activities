```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing the dataset
df = pd.read_csv('StudentsPerformance.csv')
```

```python
# First 10 rows
df.head(10)
```

```python
# Last 10 rows
df.tail(10)
```

```python
# Checking for null values
df.isnull().sum()
```

```python
# All the statistical measures of numerical features
df.describe()
```

```python
# Creating a countplot for all the categorical variables
sns.countplot(df['gender'], color='deeppink')
plt.xticks(fontsize=14)
plt.show()
```

```python
# How many males and females are there belonging from different race?
sns.countplot(x=df['race/ethnicity'], hue=df['gender'], palette='crest')
plt.show()
```

```python
# How many male and female students have opted for different types of lunch?
sns.countplot(x=df['lunch'], hue=df['gender'], palette='Paired')
plt.show()
```

```python
# Checking the distribution and skewness of quantitative fetaures
labels = ['math score', 'reading score', 'writing score']

for label in labels:
  print("Distribution of ", label)
  print("Skewness :", df[label].skew())
  sns.displot(df[label], kde=True, color='#151B54')
  plt.show()
```

```python
# Are Gender Math Score associated?
plt.scatter(x=df['gender'], y=df['math score'], color='#151B54')
plt.show()
```

```python
# Creating a heatmap to check the correlation between all features
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='mako')
```
